```python
import numpy as np
```

```python
def sigmoid (x):
    return 1/(1+np.exp(-x))
```

```python
def sigmoid_derivative(x):
    return x * (1-x)
```

```python
inputs = np.array([[0,0],[0,1],[1,0],[1,1]])
exp_output = np.array([[0],[1],[1],[0]])
```

```python
input_weights = np.random.uniform(size = (2,2))
input_bias = np.random.uniform(size = (1,2))
hidden_weights = np.random.uniform(size = (2,1))
hidden_bias = np.random.uniform(size=(1,1))
```

```python
print(input_weights)
print(input_bias)
print(hidden_weights)
print(hidden_bias)
```

```python
learning_rate = 0.1
steps = 100000
```

```python
for i in range(steps):
    input_layer_process= np.dot(inputs,input_weights) + input_bias
    
    input_layer_output= sigmoid(input_layer_process)
    
    hidden_layer_process = np.dot(input_layer_output,hidden_weights) + hidden_bias
    
    prediction_output = sigmoid(hidden_layer_process)
    
    error = exp_output-prediction_output
    Dprediction_output = error * sigmoid_derivative(prediction_output)
    error_input_layer = Dprediction_output.dot(hidden_weights.T)
    DInput_layer = error_input_layer * sigmoid_derivative(input_layer_output)
    
    hidden_weights += input_layer_output.T.dot(Dprediction_output) * learning_rate
    input_weights += inputs.T.dot(DInput_layer) * learning_rate
    hidden_bias += np.sum(Dprediction_output,axis =0,keepdims=True)*learning_rate
    input_bias += np.sum(DInput_layer,axis=0,keepdims=True)*learning_rate
    
    
    
```

```python
print(hidden_weights)
print(input_weights)
print(input_bias)
print(hidden_bias)
```
